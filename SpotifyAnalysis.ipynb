```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Spotify dataset
df = pd.read_csv('SpotifyFeatures.csv')
# Initial data exploration
print(df.head())
print(df.info())
```

```python
# EDA - Visualize data
# Example: Histogram of song popularity
plt.figure(figsize=(10, 6))
sns.histplot(df['popularity'], bins=20, kde=True)
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```

```python
# Data cleaning
#  Handle missing values
df.dropna(inplace=True)  # Remove rows with missing values
# Data analysis
# Average duration of songs by genre
avg_duration_by_genre = df.groupby('genre')['duration_ms'].mean().reset_index()

# Visualize the result
plt.figure(figsize=(12, 6))
sns.barplot(x='genre', y='duration_ms', data=avg_duration_by_genre)
plt.title('Average Duration of Songs by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Duration (ms)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Box plot of song duration by genre
#can be used to find outliers
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 2)
sns.boxplot(x='genre', y='duration_ms', data=df)
plt.title('Song Duration by Genre')
plt.xlabel('Genre')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=90)
```

```python
# Pairplot to visualize relationships between numeric features
plt.figure(figsize=(12, 8))
numeric_features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'tempo']
sns.pairplot(df[numeric_features])
plt.suptitle('Pairplot of Numeric Features', y=1.02)
```

```python
# Example 4: Heatmap to visualize correlation between numeric features
plt.figure(figsize=(12, 8))
correlation_matrix = df[numeric_features].corr()
plt.subplot(2, 2, 3)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
```

```python
# Show all plots
plt.tight_layout()
plt.show()
```
